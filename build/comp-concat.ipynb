```python
import pandas as pd 
import glob
```

```python
csv_files = glob.glob("/Users/eren/Desktop/SFND/Camera/SFND_2D_Feature_Tracking/build/Results/*.csv")
print(csv_files[0])
print(len(csv_files))
```
